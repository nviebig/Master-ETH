```python
import numpy as np
import matplotlib.pyplot as plt
from mpmath import polylog

# Range for z approaching 1 from below
z_values = np.linspace(0.001, 0.999, 500)

# Calculate g_{3/2}(z) and g_{5/2}(z) using mpmath.polylog
g_3_2_values = [polylog(1.5, z) for z in z_values]
g_5_2_values = [polylog(2.5, z) for z in z_values]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(z_values, g_3_2_values, label=r'$g_{3/2}(z)$')
plt.plot(z_values, g_5_2_values, label=r'$g_{5/2}(z)$')
plt.xlabel("z (fugacity)")
plt.ylabel("Polylogarithm Value")
plt.title("Plot of $g_{3/2}(z)$ and $g_{5/2}(z)$ approaching z = 1")
plt.legend()
plt.grid(True)
plt.show()
```
